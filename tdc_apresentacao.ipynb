```python
import requests

site = requests.get("https://norvig.com/")

site.status_code
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(site.text, 'html.parser')

soup.title
```

```python
print(soup.title.text)

print(soup.head.title)
```

```python
soup.find_all('table')
```

```python
for i in soup.find_all('a'):
    print(f"{i.text} - {i.get('href')}")
```

```python
# print(soup.find("body").find("p").find("table"))

table = soup.select("body p table")

print(table)
```

```python
rows = soup.select("body p table table tr td")
# print(rows)

print(len(rows))

for i in rows:
    print(i.select_one("b"))
```

```python
for i in soup.select('a[href^="http"]'):
    print(f"{i.text} - {i.get('href')}")
```

```python
artigo = requests.get("https://norvig.com/21-days.html")

artigo.status_code
```

```python
soup_artigo = BeautifulSoup(artigo.text, 'html.parser')

cabecalho = soup_artigo.select_one("body center")

titulo = cabecalho.select_one("h1")
autor = cabecalho.select_one("h3")

print(f"Titulo: {titulo.text} - Autor: {autor.text}")
```

```python
print(soup_artigo.select_one("table").text)
```

```python
traducoes = soup_artigo.select_one('td[valign^="top"]')
traducoes.select('a[href^="http"]')
```

```python
for t in traducoes.select('a[href^="http"]'):
    texto = t.text.strip()
    print(f"{texto} - {t.get('href')}\n |")
```

```python
rss_get = requests.get("https://norvig.com/rss-feed.xml")

rss_get.status_code
```

```python
rss = BeautifulSoup(rss_get.text, "xml")

titulo = rss.select_one("channel title")

link = rss.select_one("channel link")

descricao = rss.select_one("channel description")
```

```python
for i in rss.select("channel item"):
    print(f"Titulo: {i.title.text} \nDescrição: {i.description.text} \nLink: {i.link.text} \n--- \n")
```
